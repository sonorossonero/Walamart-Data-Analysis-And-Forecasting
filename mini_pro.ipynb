Comparision of Total Sales each store wise and year wise



Total Sales Data Yearly Wise

```python
avg_sales_2010 = data['Weekly_Sales'][data.Year==2010].groupby(data['Store']).mean()
avg_sales_2011 = data['Weekly_Sales'][data.Year==2011].groupby(data['Store']).mean()
avg_sales_2012 = data['Weekly_Sales'][data.Year==2012].groupby(data['Store']).mean()
plt.figure(figsize=(20,8))
sns.lineplot(x=avg_sales_2010.index ,y=avg_sales_2010.values)
sns.lineplot(x=avg_sales_2011.index ,y=avg_sales_2011.values)
sns.lineplot(x=avg_sales_2012.index ,y=avg_sales_2012.values)
plt.grid()
plt.xticks(np.arange(1,45,step=1))
plt.legend(['2010','2011','2012'],loc = 'best' , fontsize=16)
plt.title("Yearly Average Sales Data " , fontsize=18)
plt.xlabel("Store" , fontsize=16)
plt.ylabel('Average Sales' , fontsize=16)
plt.show()
```
